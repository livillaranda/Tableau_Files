```python
# Dependencies
import pandas as pd
import numpy as np
from scipy import stats
```

```python
# Load Dataset (Primary - 2020)
bike = pd.read_csv("../../../202007-citibike-tripdata.csv")
```

```python
# Display Data
c_bike = pd.DataFrame(bike)
c_bike.head(5)
```

```python
# Rename Columns 
c_bike = c_bike.rename(columns={"tripduration":"Trip", "starttime": "Start", "stoptime": "Stop", "start station id": "Start ID", 
                                "start station name": "Start Station", "start station latitude": "Start Lat", "start station longitude": "Start Long",
                                "end station id": "End ID", "end station name": "End Station", "end station latitude": "End Lat",
                                "end station longitude": "End Long", "bikeid": "Bike ID", "usertype": "User", "birth year": "BY", "gender": "Gender"})
c_bike.head(5)
```

```python
# Load Dataset (Comparison - 2019)
bike19 = pd.read_csv("../../../201907-citibike-tripdata.csv")
```

```python
# Display Data (2019)
c_bike19 = pd.DataFrame(bike19)
c_bike19.head(5)
```

```python
# Rename Columns 
c_bike19 = c_bike19.rename(columns={"tripduration":"Trip", "starttime": "Start", "stoptime": "Stop", "start station id": "Start ID", 
                                "start station name": "Start Station", "start station latitude": "Start Lat", "start station longitude": "Start Long",
                                "end station id": "End ID", "end station name": "End Station", "end station latitude": "End Lat",
                                "end station longitude": "End Long", "bikeid": "Bike ID", "usertype": "User", "birth year": "BY", "gender": "Gender"})
c_bike19.head(5)
```

```python
# Trip Duration (2020)
avg20 = c_bike["Trip"].mean()
print("Average is", avg20, "seconds.")

min20 = c_bike["Trip"].min()
print("Min is", min20, "seconds.")

max20 = c_bike["Trip"].max()
print("Max is", max20, "seconds.")
```

```python
# Trip Duration (2019)
avg19 = c_bike19["Trip"].mean()
print("Average is", avg19, "seconds.")

min19 = c_bike19["Trip"].min()
print("Min is", min19, "seconds.")

max19 = c_bike19["Trip"].max()
print("Max is", max19, "seconds.")
```

```python
# Most Frequently Used Bike
bikeid20 = stats.mode(c_bike["Bike ID"])
print("Most used bike is", bikeid20, "in July 2020.")

bikeid19 = stats.mode(c_bike19["Bike ID"])
print("Most used bike is", bikeid19, "in July 2019.")
```
